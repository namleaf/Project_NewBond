```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from openpyxl import load_workbook

file = 'WHO.xlsx'
baby_data = 'baby.xlsx' 

# Load spreadsheet: xl
xl = pd.ExcelFile(file)

#load workbook
book = load_workbook(baby_data)
writer = pd.ExcelWriter(baby_data, engine = 'openpyxl')
writer.book = book

#gender of the newborn 
gender1 = 'Mädchen'
gender2 = 'Junge'

gender = input('Geschlecht:\n')
name = input('Unser neues Familienmitlgied heißt \n')

if gender == gender1 or gender == 'Mädchen':

    # Load a sheet into a DataFrame by name: df1
    df1 = xl.parse('Größe Mädchen (cm)')
    df2 = xl.parse('Gewicht Mädchen (gramm)')

    # create copy to avoid data manipulation
    girl_length = df1.copy()
    girl_weight = df2.copy()

    #myBaby
    i=0
    length_of_my_baby_girl = []
    weight_of_my_baby_girl = []
    while i<=13:
        if i == 0:
            #enter length
            length_girl_week = input('Mein Baby war \n cm Groß bei der Geburt')
            length_of_my_baby_girl.append(length_girl_week)

            #enter weight
            weight_girl_week = input('Mein Baby wog \n Gramm bei der Geburt')
            weight_of_my_baby_girl.append(weight_girl_week)

            i = i+1

        elif i>0 and i<=13:
            #enter length
            print("""\nHallo, es ist die {}. Woche und somit Zeit für einen neuen Eintrag!""".format(i))
            length_girl_week = input('Mein Baby ist jetzt \n cm Groß')

            length_of_my_baby_girl.append(length_girl_week)

            #enter weight
            print("""\nNeben dem Größe fehlt nur noch das Gewicht für die {}. Woche.""".format(i))
            weight_girl_week = input('Mein Baby wiegt \n Gramm')
            weight_of_my_baby_girl.append(weight_girl_week)

            i = i+1

        else:
            print('Ups. Da ist wohl ein Fehler passiert. Bitte geben Sie das Geschlecht Ihres Kindes an.\n')
 


    #merge name of baby and length/weight
    girl_length[name] = length_of_my_baby_girl
    girl_weight[name] = weight_of_my_baby_girl

    #convert object into float64
    girl_length[name] = pd.to_numeric(girl_length[name])
    girl_weight[name] = pd.to_numeric(girl_weight[name])

    #drop NaN
    girl_length[name] = girl_length[name].dropna()
    girl_weight[name] = girl_weight[name].dropna()

    # diagramm data
    plt.figure(0).canvas.set_window_title('Größe  Mädchen')

    plt.plot(girl_length.Woche, girl_length[name],'o', color='black', label=name)
    plt.plot(girl_length.Woche, girl_length['M'], color='green', label='Mittelwert')
    plt.plot(girl_length.Woche, girl_length['1 SD'], color='yellow', label='1. Standartabweichung')
    plt.plot(girl_length.Woche, girl_length['2 SD'], color='red', label='2. Standartabweichung')
    plt.plot(girl_length.Woche, girl_length['-1 SD'], color='yellow')
    plt.plot(girl_length.Woche, girl_length['-2 SD'], color='red')


    # customization of plot
    plt.title('Größe Mädchen (cm)')
    plt.xlabel('Zeit (Woche)')
    plt.ylabel('Größe (cm)')
    plt.axis((0, 14, 42, 68))


    # legend
    plt.legend(loc='lower right')

    #save pic
    plt.savefig("""Length_{}.png""".format(name))

    #repeat for weight
    plt.figure(1).canvas.set_window_title('Gewicht Mädchen')

    plt.plot(girl_weight.Woche, girl_weight[name],'o', color='black', label=name)
    plt.plot(girl_weight.Woche, girl_weight['M'], color='green', label='Mittelwert')
    plt.plot(girl_weight.Woche, girl_weight['1 SD'], color='yellow', label='1. Standartabwechung')
    plt.plot(girl_weight.Woche, girl_weight['2 SD'], color='red', label='2. Standartabweichung')
    plt.plot(girl_weight.Woche, girl_weight['-1 SD'], color='yellow')
    plt.plot(girl_weight.Woche, girl_weight['-2 SD'], color='red')

    plt.title('Gewicht Mädchen (gramm)')
    plt.xlabel('Zeit (Woche)')
    plt.ylabel('Gewicht (Gramm)')
    plt.axis((0, 14, 3000, 6000))

    # legend
    plt.legend(loc='lower right')

    #save pic
    plt.savefig("""Weight_{}.png""".format(name))

    #add data to workbook
    entry = {'Length':length_of_my_baby_girl, 'Weight': weight_of_my_baby_girl}
    df = pd.DataFrame(entry)
    df.to_excel(writer, sheet_name = name)
    writer.save()
    writer.close()

    plt.show()


elif gender == gender2 or gender == 'Junge':

    # Boy

    # Load a sheet into a DataFrame by name: df1
    df4 = xl.parse('Größe Jungen (cm)')
    df5 = xl.parse('Gewicht Jungen (gramm)')

    boy_length = df4.copy()
    boy_weight = df5.copy()


    #myBaby
    i=0
    length_of_my_baby_boy = []
    weight_of_my_baby_boy = []
    while i<=13:
        if i == 0:
            #enter length
            length_boy_week = input('Mein Baby war \n cm Groß bei der Geburt')
            length_of_my_baby_boy.append(length_boy_week)

            #enter weight
            weight_boy_week = input('Mein Baby wog \n Gramm bei der Geburt')
            weight_of_my_baby_boy.append(weight_boy_week)

            i = i+1

        elif i>0 and i<=13:
            #enter length
            print("""\nHallo, es ist die {}. Woche und somit Zeit für einen neuen Eintrag!""".format(i))
            length_boy_week = input('Mein Baby ist \n cm Groß in der  Woche')
            length_of_my_baby_boy.append(length_boy_week)

            #enter weight
            print("""\nNeben dem Größe fehlt nur noch das Gewicht für die {}. Woche.""".format(i))
            weight_boy_week = input('Mein Baby wiegt \n Gramm')
            weight_of_my_baby_boy.append(weight_boy_week)

            i = i+1

        else:
            print('Ups. Da ist wohl ein Fehler passiert. Bitte geben Sie das Geschlecht Ihres Kindes an.\n')
 


    #merge name of baby and length/weight
    boy_length[name] = length_of_my_baby_boy
    boy_weight[name] = weight_of_my_baby_boy

    #convert object into float64
    boy_length[name] = pd.to_numeric(boy_length[name])
    boy_weight[name] = pd.to_numeric(boy_weight[name])

    #drop NaN
    boy_length[name] = boy_length[name].dropna()
    boy_weight[name] = boy_weight[name].dropna()

    # diagramm data

    boy_fig = plt.figure(3).canvas.set_window_title('Größe Jungen')

    plt.plot(boy_length.Woche, boy_length[name],'o', color='black', label=name)
    plt.plot(boy_length.Woche, boy_length['M'], color='green', label='Mittelwert')
    plt.plot(boy_length.Woche, boy_length['1 SD'], color='yellow', label='1. Standartabweichung')
    plt.plot(boy_length.Woche, boy_length['2 SD'], color='red', label='2. Standartabweichung')
    plt.plot(boy_length.Woche, boy_length['-1 SD'], color='yellow')
    plt.plot(boy_length.Woche, boy_length['-2 SD'], color='red')

    # customization of plot
    plt.title('Größe Junge (cm)')
    plt.xlabel('Zeit (Woche)')
    plt.ylabel('Größe (cm)')
    plt.axis((0, 14, 42, 68))

    # legend
    plt.legend(loc='lower right')

    #save pic
    plt.savefig("""Length_{}.png""".format(name))

    # diagramm data
    plt.figure(4).canvas.set_window_title('Gewicht Jungen')

    plt.plot(boy_weight.Woche, boy_weight[name],'o', color='black', label=name)
    plt.plot(boy_weight.Woche, boy_weight['M'], color='green', label='Mittelwert')
    plt.plot(boy_weight.Woche, boy_weight['1 SD'], color='yellow', label='1. Standartabweichung')
    plt.plot(boy_weight.Woche, boy_weight['2 SD'], color='red', label='2. Standartabweichung')
    plt.plot(boy_weight.Woche, boy_weight['-1 SD'], color='yellow')
    plt.plot(boy_weight.Woche, boy_weight['-2 SD'], color='red')

    plt.title('Gewicht Junge (gramm)')
    plt.xlabel('Zeit (Woche)')
    plt.ylabel('Gewicht (Gramm)')
    plt.axis((0, 14, 3000, 6000))

    # legend
    plt.legend(loc='lower right')

    #save pic
    plt.savefig("""Weight_{}.png""".format(name))

    #add data to workbook
    entry = {'Length':length_of_my_baby_boy, 'Weight': weight_of_my_baby_boy}
    df = pd.DataFrame(entry)
    df.to_excel(writer, sheet_name = name)
    writer.save()
    writer.close()
    plt.show()
else:
     print('Ups. Da ist wohl ein Fehler passiert. Bitte geben Sie das Geschlecht Ihres Kindes an.\n')

```
